```python
# Analysis of Variance(ANOVA)

from scipy.stats import f_oneway
cat_col=['Brand','Model','Fuel_Type','Transmission','Owner_Type']
for feature in cat_col:
    groups=[data[data[feature]==group]['Price'] for group in data[feature].unique()]
    f_stat,p_value=f_oneway(*groups)

    print(f"ANOVA result for{feature}:")
    print(f"F-statistics: {f_stat:2f}")
    print(f"P-value: {p_value:4f}")
```

```python
from sklearn.preprocessing import StandardScaler
numerical_features = ['Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats', 'Car_Age']
scaler = StandardScaler()
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
x= data.drop(columns=['Price'])
y= data['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
y_test_predlr=lr_model.predict(x_test) #for evaluating test set
y_train_predlr=lr_model.predict(x_train) #for evaluating train set
# Evaluation Metrics
lr_metrics = {
            'Test Mean Squared Error':mean_squared_error(y_test,y_test_predlr),
            'Test Root Mean Squared Error':np.sqrt(mean_squared_error(y_test,y_test_predlr)),
            'Test Mean Absolute Error':mean_absolute_error(y_test,y_test_predlr),
            'Test R-Squared':r2_score(y_test,y_test_predlr),
            'Train Mean Squared Error':mean_squared_error(y_train,y_train_predlr),
            'Train Root Mean Squared Error':np.sqrt(mean_squared_error(y_train,y_train_predlr)),
            'Train Mean Absolute Error':mean_absolute_error(y_train,y_train_predlr),
            'Train R-Squared Error':r2_score(y_train,y_train_predlr)
        }                                                   
print(lr_metrics)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
xgb_grid_model = XGBRegressor(colsample_bytree=0.8,learning_rate=0.05,max_depth=7,min_child_weight=3,n_estimators=100,subsample=0.8,random_state=42)
xgb_grid_model.fit(x_train, y_train)
xgb_random_model = XGBRegressor(colsample_bytree=0.9,learning_rate=0.1578,max_depth=3,min_child_weight=5,n_estimators=200,subsample=0.8,random_state=42)
xgb_random_model.fit(x_train, y_train)
for model, name in zip([xgb_grid_model, xgb_random_model], ["GridSearchCV", "RandomizedSearchCV"]):
    y_test_predlr = model.predict(x_test)
    mae = mean_absolute_error(y_test, y_test_predlr)
    mse = mean_squared_error(y_test, y_test_predlr)
    r2 = r2_score(y_test, y_test_predlr)
    print(f"{name} Model - MAE: {mae}, MSE: {mse}, R: {r2}")
```

```python
import shap
explainer=shap.Explainer(xgb_random_model,x_test)
shap_values=explainer(x_test)
shap.summary_plot(shap_values,x_test)
```
